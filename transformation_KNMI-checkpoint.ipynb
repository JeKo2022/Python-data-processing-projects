```python
import csv 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn as sk # for linear regression (fitting straight line to data, etc.)
from sklearn.linear_model import LinearRegression
```

```python
## LOADING THE DATA: correctly read input files into pandas DataFrame

# input all filenames
filenames = ['tas_1901_1930_NLD.csv', 'tas_1931_1960_NLD.csv', 'tas_1961_1990_NLD.csv', 'tas_1991_2016_NLD.csv']

# read al files and append to list for concatenation
all_dfs = []
for filename in filenames:
    
    # do not use the first column as the index, set header to first row, skip spaces after delimiter
    df = pd.read_csv(f'{filename}', index_col = False, header = 0, skipinitialspace = True)
    all_dfs.append(df)

# concatenate all dataframes into one large df
concat_df = pd.concat(all_dfs, ignore_index=True)
display(concat_df)
```

```python
## PLOTTING THE DATA AND THE TREND

# select data 
years = concat_df["Year"]
temps = concat_df["Temperature - (Celsius)"]

# make a basic plot of the selected data. 
plt.plot(years, temps)
plt.xlabel("years")
plt.ylabel("temperatures")
```

```python
## FITTING A LINE TO THE DATA
def plots(year_column, temperature_column, input_df):
    ''' 
    Function fits a line to the data and also plots the fitted line.
    '''

    # aking a new instance of the model 
    model = LinearRegression()

    # convert the years series to a numpy array and  reshape to 2D
    y = temperature_column
    x = year_column
    x = x.to_numpy()
    
    # reshape to lengths of "Year" column
    x = x.reshape((year_column.shape[0], 1))

    # call fit to change the model to fit the data
    model.fit(x, y)

    # call predict to return the fitted results for each of those years
    model.predict(x)

    # add acolumn to dataframe containing the results of fitting the linear model to the temperatures 
    input_df["predicted"] = model.predict(x)

    # plot fitted linear line in addition to the original data and finally label both lines
    years = year_column
    predicted = input_df["predicted"]
    temps = temperature_column

    # plot
    plt.plot(years, temps, label = "temperatures")
    plt.plot(years, predicted, label = "predicted temperatures")
    plt.xlabel("years")
    plt.ylabel("temperatures")

    # places the legend at the location with the minimum overlap
    plt.legend(loc = "best")
    
    plt.show()
    
    return years, predicted, temps
    
    
# Plot year averages and linear fitted line for the combined data using functions from before.
years, predicted, temps = plots (concat_df["Year"], concat_df["Temperature - (Celsius)"], concat_df)
    
    
```

```python
def plot_months (month, month_year, month_temps):
    '''
    This function plots the temperatures for a given month.
    Note that we use a copy of the concat_df, so we don't change concat_df itself.
    '''
    
    # aking a new instance of the model 
    model = LinearRegression()
    
    # convert the years series to a numpy array and  reshape to 2D
    y = month_temps
    x = month_year
    x = x.to_numpy()
    
    # reshape to lengths om "Year" column
    x = x.reshape((month_year.shape[0], 1))

    # call fit to change the model to fit the data
    model.fit(x, y)

    # call predict to eturn the fitted results for each of those years
    model.predict(x)

    # add acolumn to dataframe containing the results of fitting the linear model to the temperatures.
    month["predicted"] = model.predict(x)

    # plot fitted linear line in addition to the original data and finally label both lines
    years = month_year
    predicted = month["predicted"]
    temps = month_temps
    
    # plot specific temperatures and fitted line for month
    plt.plot(years, temps, label = f"temperatures {month_name}")
    plt.plot(years, predicted, label = f"predicted temperatures {month_name}")
    plt.xlabel("years")
    plt.ylabel("temperatures")
    
    # places the legend at the location with the minimum overlap
    plt.legend(loc = "best")
    
    # plt.show is used to plot the figures seperately
    plt.show()
    
    # despite plt.show, return is necessary to plot next plots
    return years, predicted, temps

## SELECTING MONTHS
months = ["Jun", "Jul", "Aug"]

for month_name in months:
    
    # Use Boolean indexing to filter out a specific month and select specific month
    month = concat_df[concat_df["Statistics"] == f"{month_name} Average"].copy()
    month_year = month["Year"]
    month_temps = month["Temperature - (Celsius)"]

    # extract the temperatures and fitted data for month
    years, predicted, temps = plot_months (month, month_year, month_temps)
    
    
```

```python
## AVERAGES OVER LARGER TIME PERIODS
   
# compute the average for the whole year, combining each of the months
mean_df = concat_df.copy()
mean_years = mean_df.groupby(["Year"], as_index = False).mean()

# plot year averages and linear fitted line for the combined data using functions from before
years, predicted, temps = plots(mean_years["Year"], mean_years["Temperature - (Celsius)"], mean_years)
```

```python
# compute the average for the whole year, combining each of the months 
df = concat_df.copy()

# compute the average for the decades
df["decades"] = df["Year"]//10*10
mean_decades = df.groupby(["decades"], as_index = False).mean()

# plot decade averages and linear fitted line for the combined data using functions from before
years, predicted, temps = plots(mean_decades["decades"], mean_decades["Temperature - (Celsius)"], mean_decades)
    
    
```
